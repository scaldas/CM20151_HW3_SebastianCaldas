```python
file_name = 'Houston-security.xls'
df = pd.read_excel(file_name, 0)
mask = df['Offense Type'] == 'Aggravated Assault'
df_masked = df[mask]
hourly = df_masked.groupby('Hour').sum()
print('La hora del dia mas segura segun el delito de asalto agravado es a las ' + str(hourly.idxmin( )[0]) + ' pues presenta el menor numero de incidentes (con apenas '+ str(hourly.min( )[0]) +')')
```
