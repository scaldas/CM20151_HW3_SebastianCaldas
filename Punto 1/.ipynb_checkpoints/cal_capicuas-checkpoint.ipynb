```python
capicuas = []
for j in range(0,10):
    for i in range(1000,10000):
        rvr = str(i)[::-1]
        capicuas.append(int(str(i) + str(j) + rvr))
print('Se imprimen ' + str(len(capicuas)) + ' numeros capicua:') 
for capi in capicuas:
    print(capi)
```

```python
capicuas_bin = []
capicuas_bin2 = []

for capi in capicuas:
    num_bin = bin(capi)[2:]
    num_rever = num_bin[::-1]
    if num_bin == num_rever:
        capicuas_bin.append(int(num_bin))
        capicuas_bin2.append(capi) 
print('Se imprimen ' + str(len(capicuas_bin)) + ' numeros capicua: ')

i = 0
for number in capicuas_bin:
    print(str(capicuas_bin2[i]) + ' = ' + str(number))
    i = i + 1
```

```python
my_str = input("Ingresa el string: ")
import unicodedata 
my_str = ''.join((c for c in unicodedata.normalize('NFD', my_str) if unicodedata.category(c) != 'Mn'))
punctuations = ''' …!¡()-[]{};:'"\,<>./¿?@#$%^&*_~'''
no_punct = "" 
for char in my_str: 
    if char not in punctuations: 
        no_punct = no_punct + char 
no_punct = no_punct.casefold()
rev_str = reversed(no_punct)
print(no_punct)
if list(no_punct) == list(rev_str):
   print("Es palindroma")
else:
   print("No es palindroma")
```
